## Tratamento de Dados Faltantes

```python
import pandas as pd
```

```python
dados = pd.read_csv("dados/aluguel_residenciais.csv", sep = ";")
```

```python
dados.isnull() #ele cria um DataFrame e marca TRUE se o dado for nulo
```

```python
dados.notnull() #ele cria um DataFrame e marca FALSE se o dado for nulo
```

```python
dados.info() #usado para analizar quais colunas tem valores NULOS!!
```

```python
dados[dados["Valor"].isnull()] #ele pega todos os valores nulos da coluna "Valor"
```

```python
dados[dados["IPTU"].isnull()] #ele pega todos os valores nulos da coluna "IPTU"
```

```python
dados[dados["Condominio"].isnull()] #ele pega todos os valores nulos da coluna "Condominio"
```

```python
#Nos sabemos pelo dados.info() que existem 9 valores na coluna "Valor" que sao NaN
#
#para retirar esses 9 valores utilizamos o dropna() e informamos a coluna que gostaria de eliminar os valores NaN

A = dados.shape[0] #mostrando a quantidade de registros antes do DROPNA()

dados.dropna(subset = ["Valor"], inplace = True) #fazendo o drop

B = dados.shape[0]

A - B #Mostrando a diferenca entre antes e depois do DROPNA() na coluna VALOR


```

```python
selecao = (dados["Tipo"] == "Apartamento") & (dados["Condominio"].isnull())
```

```python
A = dados.shape[0] 

dados = dados[~selecao] #o TIO INVERTE A SELECAO (WOOOOW)

B = dados.shape[0]

A - B 
```

```python
dados.fillna({"Condominio": 0, "IPTU": 0}) #o FILLNA() substitui todos os valores NaN pelo valor que quiser!!
```

```python
dados.to_csv("dados/aluguel_residenciais.csv", sep = ";", index = False)
```
